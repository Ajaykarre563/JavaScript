# **How to display your output in javascript?**

console.log() is a JavaScript function used to print messages on the screen (usually in the browser's console).

It helps show values of variables or check if your code is working.

It takes things like arrays, objects, or simple messages as input.

It returns whatever you tell it to print.

# **Output:**

You can use console.log() to write or log messages.

It helps to show what the code is doing or to alert if something goes wrong.

# **Syntax:**



```
console.log("Text to show");

```



# **Example:**



```
console.log('hello World');

```

```
// Using var
var x = 5;
var x = 10; // ✅ Allowed: var can be re-declared and updated
console.log(x); // Output: 10

// Using let
let y = 5;
// let y = 10; ❌ Error: can't re-declare a let variable in the same scope
y = 15; // ✅ Allowed: let can be updated
console.log(y); // Output: 15

// Using const
const z = 5;
// z = 10; ❌ Error: can't re-assign a const variable
console.log(z); // Output: 5

```



# **Examples DataTypes:**



```
// Number
var age = 25;
console.log("Number:", age); // Output: Number: 25

// String
var name = "Tom";
console.log("String:", name); // Output: String: Tom

// Boolean
var isStudent = true;
console.log("Boolean:", isStudent); // Output: Boolean: true

// Object
var person = { name: "Tom", age: 25 };
console.log("Object:", person); // Output: Object: {name: "Tom", age: 25}

// Array
var fruits = ["apple", "banana", "cherry"];
console.log("Array:", fruits); // Output: Array: ["apple", "banana", "cherry"]

// Null
var emptyValue = null;
console.log("Null:", emptyValue); // Output: Null: null

// Undefined
var notAssigned;
console.log("Undefined:", notAssigned); // Output: Undefined: undefined

```

# **Examples:**



```
// Primitive (Passed by Value)
let a = 10;
let b = a;  // Copies the value
b = 20;
console.log(a);  // Output: 10
console.log(b);  // Output: 20

// Non-Primitive (Passed by Reference)
let arr1 = [1, 2, 3];
let arr2 = arr1;  // Copies the reference (both point to the same array)
arr2.push(4);
console.log(arr1);  // Output: [1, 2, 3, 4]
console.log(arr2);  // Output: [1, 2, 3, 4]

```
