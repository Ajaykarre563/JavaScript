# **What is a JavaScript Object?**

**Objects are a way to group multiple values together into a single variable.**

**They consist of key-value pairs (also known as name-value pairs), where:**

The key is a string (e.g., "name", "age")

The value can be anything: string, number, boolean, array, or even another object.



# **Structure:**



```
var person = {
  name: "John",
  age: 31,
  favColor: "green",
  height: 183
};

```



image.png



```
class Person {
    constructor(name, age) {
        this.name = name;
        this.age = age;
    }
}

const p1 = new Person("RAM", 15);
const p2 = new Person("laxman", 13);

console.log(p1.name); // RAM
console.log(p1.age);  // 15
console.log(p2.name); // laxman
console.log(p2.age);  // 13

```



# **What is a Method?**

A method is just a function that is stored inside an object. You can use it to define behavior for that object.

# **Syntax**

**Creating a method inside an object:**



```
methodName = function() {
  // code here
}

```



**Accessing (calling) a method:**



```
objectName.methodName();

```



# **Important Note:**

A method belongs to an object.

Use the this keyword inside the method to refer to the current object.

This lets you access other properties or methods of the same object.





```
let person = {
  name: "Alice",
  greet: function() {
    console.log("Hello, " + this.name);
  }
};

person.greet();  // Output: Hello, Alice

```



# **What is a Constructor Function?**

A constructor function is used to create multiple objects with the same structure and behavior.



```
<!DOCTYPE html>
<html>
<head>
  <title>JavaScript Method Example</title>
</head>
<body>

<script>
  // Constructor function
  function person(name, age) {
    this.name = name;
    this.age = age;

    // Defining a method inside the constructor
    this.changeName = function(name) {
      this.name = name;
    }
  }

  // Creating an object
  var p = new person("David", 21);

  // Calling the method to change the name
  p.changeName("John");

  // Displaying the updated name
  document.write(p.name);  // Output: John
</script>

</body>
</html>

```



# **Method Call:**

# **Math Object in JavaScript**
The Math object allows you to perform mathematical tasks and provides several properties and methods.

Unlike most objects, Math has no constructor  you cannot create instances of it.

It is a static object, so:

All methods and properties are accessed directly using Math, without creating an object.



```
Math.method(number)

```





```
Math.sqrt(16);  // returns 4
Math.random();  // returns a random number between 0 and 1


```
