# **Type conversion is the process of converting data from one type to another.**

**Example**: Converting a string like "123" into a number like 123.

# **In JavaScript, there are two types of type conversion:**

**Implicit Conversion**  Also called automatic type conversion, done by JavaScript itself.

**Explicit Conversion**  Also called manual type conversion, done by the developer using functions or methods.

# **Implicit Conversion to String**

Happens when a numeric value is added (+) to a string.

JavaScript converts the number to a string and performs string concatenation.



```
let result;
result = '3' + 2;
console.log(result); // "32"

```



# **Implicit Conversion to Number**

Occurs when a numeric string is used with arithmetic operators like -, /, or *.

JavaScript converts the strings to numbers before performing the operation.



```
let result;
result = '4' - '2';
console.log(result); // 2

```



# **Implicit Boolean Conversion to Number**

**When Boolean values are used in a numeric operation:**

true becomes 1

false becomes 0



```
let result;
result = '4' - true;
console.log(result); // 3

```



# **Explicit conversion:**

**Convert one datatype to another datatype by using bulit-in functions it is mainly used for manipulate operations.**

# **Note:**

**You can generate numbers from strings using:**

**parseInt():** Converts a string to an integer.

**parseFloat():** Converts a string to a floating-point number.

Unary operator +: Converts a string to a number.

**Math.floor():** Rounds a number down to the nearest integer.



```
let result = '20.1';
console.log(parseInt(result)); // 20
console.log(parseFloat(result)); // 20.01
console.log(Math.floor(result)); // 20
```

# **string Datatype:**

# **Key Points:**

A string is any text wrapped in single (') or double quotes (").

JavaScript strings are used for storing and manipulating text.

String is a reference type datatype (meaning it's an object under the hood).



```
let a = "car"; // Double quotes
let b = 'bus'; // Single quotes
```



# **Key Points:**

**length:** Returns the length of a string.

**charAt():** Returns the character at a specified index.

**indexOf():** Returns the index of the first occurrence of a value in a string.

**toUpperCase():** Converts a string to uppercase letters.

**toLowerCase():** Converts a string to lowercase letters.

**toString():** Returns a string or string object as a string.



```
let str = "Hello World";
console.log(str.length);         // 11
console.log(str.charAt(1));      // "e"
console.log(str.indexOf("o"));   // 4
console.log(str.toUpperCase());  // "HELLO WORLD"
console.log(str.toLowerCase());  // "hello world"
console.log(str.toString());     // "Hello World"
```

# **escape characters**

# **Key Points:**

Strings must be written within quotes, but including quotes or other special characters directly can cause JavaScript to misinterpret the string.

Example: let text = "We are the so-called "Vikings" from the north."; will cause an error because the inner double quotes conflict with the outer ones.

Solution: Use the backslash (\) escape character to turn special characters into string characters.



```
let text = "We are the so-called \'Vikings\' from the north.";
console.log(text)
```
