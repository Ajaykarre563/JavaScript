# **DOM Events:**

**The HTML DOM allows JavaScript to react to HTML events.**

**A JavaScript function can be executed when an event occurs, such as a user clicking on an HTML element.**

**To execute code when a user interacts with an element (like a click), JavaScript code is added to an HTML event attribute (e.g., onclick, onmouseover, etc.).**

```
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>Events</title>
  <style>
    body {
      font-family: Arial, sans-serif;
      padding: 30px;
    }
    input {
      padding: 8px;
      margin-bottom: 10px;
    }
    button {
      padding: 20px;
      margin: 10px 10px 0 0;
      border: none;
      cursor: pointer;
      font-size: 16px;
    }
    .green-btn {
      background-color: green;
      color: white;
    }
    .black-btn {
      background-color: black;
      color: white;
    }
  </style>
</head>
<body>

  <h2>JavaScript HTML Events</h2>

  <label for="fname">Enter your name:</label>
  <input type="text" id="fname" onchange="upperCase()">
  
  <p>When you leave the input field, a function is triggered which transforms the input text to upper case.</p>

  <!-- Green Thank You Button -->
  <button class="green-btn">Thank You</button>

  <!-- Black Click Me Button with onclick -->
  <button class="black-btn" onclick="clickedMessage()">Click Me</button>

  <script>
    function upperCase() {
      const x = document.getElementById("fname");
      x.value = x.value.toUpperCase();
    }

    function clickedMessage() {
      alert("Button clicked!");
    }
  </script>

</body>
</html>

```
