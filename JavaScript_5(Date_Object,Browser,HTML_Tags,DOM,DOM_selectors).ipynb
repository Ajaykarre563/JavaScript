# **What is the DOM?**

When a webpage is opened in a browser, the HTML is loaded and rendered on the screen.

The browser creates a Document Object Model (DOM) — a structured, object-oriented representation of the page.



# **Why is DOM Important?**
JavaScript uses the DOM to:

Add elements

Delete elements

Modify elements

— all dynamically without reloading the page.

# **🧠 Concept:**

**All HTML elements are objects.**

The document object in JavaScript provides methods to access those HTML elements.

# **📌 Common Methods for Selecting Elements:**

# **By ID**


```
document.getElementById(id)

```

Returns the element with the specified id.

Only one element per ID should exist on a page.



# **By Class Name**



```
document.getElementsByClassName(name)

```
Returns a collection (HTMLCollection) of all elements with the specified class name.




# **By Tag Name**

```
document.getElementsByTagName(name)

```

Returns a collection of all elements with the given tag name (like div, p, a, etc.).



# **Example:**



```
<!DOCTYPE html>
<html>
<head>
  <title>Selecting Elements</title>
</head>
<body>

  <h1 id="title">Hello, World!</h1>
  <p class="info">This is a paragraph.</p>
  <p class="info">This is another paragraph.</p>
  <div>This is a <strong>div</strong> element.</div>

  <script src="script.js"></script>
</body>
</html>

```





```
// 1. Selecting by ID
const titleElement = document.getElementById("title");
console.log("By ID:", titleElement.textContent);

// 2. Selecting by class name
const infoElements = document.getElementsByClassName("info");
for (let i = 0; i < infoElements.length; i++) {
  console.log("By Class:", infoElements[i].textContent);
}

// 3. Selecting by tag name
const divElements = document.getElementsByTagName("div");
for (let i = 0; i < divElements.length; i++) {
  console.log("By Tag:", divElements[i].textContent);
}

```
